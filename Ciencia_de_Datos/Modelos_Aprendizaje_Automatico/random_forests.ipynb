```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=5000, n_features=20, random_state=3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)
```

```python
precision = random_forest.score(X_test, y_test) #precision con test data
print('La precision es: ', precision*100)
```

```python

y_pred = random_forest.predict(X_test) #prediccion con test data
matriz_confusion = confusion_matrix(y_test, y_pred) #matriz de confucion con datos de prueba "y" y x

plt.subplots(figsize = (4,4)) 
sns.heatmap(matriz_confusion, annot = True, fmt = ".0f")
plt.xlabel("Prediccion")
plt.ylabel("Actual")
plt.show()
```
