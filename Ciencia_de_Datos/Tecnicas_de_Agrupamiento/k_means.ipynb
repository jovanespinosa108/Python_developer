```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("clusters.csv", delimiter=";")
```

```python
print(df.head(10))
```

```python
kmeans = KMeans(n_clusters=4) #KMeans divide los datos en 4 clusters
kmeans.fit(df) #entrena el modelo
y_kmeans = kmeans.predict(df) #hace predicciones de los 4 clusters
cluster_centers = kmeans.cluster_centers_ #almacena en arrays los centroides
```

```python
plt.scatter(df["Latitud"], df["Longitud"], c=y_kmeans, s=20) #imprime los clusters
plt.scatter(cluster_centers[:,0], cluster_centers[:, 1], c='red', s=200) #imprime centroides en rojo
plt.xlabel("Latitud") #asigna nombre al eje x
plt.ylabel("Longitud") #asigna nombre al eje y
```
