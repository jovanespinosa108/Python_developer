```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("clientes.csv", delimiter=";")
```

```python
print(df.head())
```

```python
dendrogram(linkage(df, method="ward", metric="euclidean")) #grafica el dendograma
plt.axhline(y=15000, color='r', linestyle="--") #crea linea de estilo punteada roja en 15k,
```

```python
#crea agrupacion aglomerativa con 3 clusters,
agglomerative = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage="ward")
agglomerative.fit(df)
lables = agglomerative.labels_ #obtiene los lables
```

```python
print(lables) #cada numero representa un lable o cluster
```

```python
plt.scatter(df['Salario'], df['Gasto'], c=lables) #imprime las columnas
plt.xlabel("Salario Mensual") #asigna nombre a axis x
plt.ylabel("Porcentaje Gasto Mensual") #asigna numbre a axis y
```
